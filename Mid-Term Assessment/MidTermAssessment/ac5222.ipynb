````python
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score, accuracy_score, roc_curve


X_train, X_test, y_train, y_test = train_test_split(coord, label, test_size = 0.2,random_state = 3) # Use training and test sets from Q1

# Define hyperparameters

list_of_max_depth = [5, 10, 50]
list_of_max_leaves = [2, 4, 5, 10]
list_of_learning_rate = [0.001, 0.01, 0.1, 0.3]
list_of_n_estimators = [50, 500]

list_of_hparams = [(d, l, lr, n) for d in list_of_max_depth 
                                 for l in list_of_max_leaves 
                                 for lr in list_of_learning_rate 
                                 for n in list_of_n_estimators]

best_acc_xg = 0
best_roc_xg = 0

results_df = pd.DataFrame(columns=['XGBoost','max_depth', 'n_estimators', 'max_leaves', 'learning_rate', 'accuracy', 'roc_score'])


for max_depth, max_leaves, learning_rate, n_estimators in list_of_hparams:
    # Define and fit your model 
    clf = XGBClassifier(max_depth = max_depth, 
                        max_leaves = max_leaves, 
                        learning_rate = learning_rate, 
                        n_estimators = n_estimators,
                        random_state = 1)
    # Fit model
    clf.fit(X_train, y_train)
    # Calculate performance metrics
    acc = accuracy_score(y_test, clf.predict(X_test))
    roc = roc_auc_score(y_test, clf.predict_proba(X_test)[:,1])
    if roc > best_roc_xg:
        best_roc_xg = roc
        best_acc_xg = acc
        with open('best_mnist_xg.pkl','wb') as f:
            pickle.dump(clf, f)
    elif roc==best_roc_xg and acc > best_acc_xg:
        best_acc_xg = acc
        with open('best_mnist_xg.pkl','wb') as f:
            pickle.dump(clf, f)
    model_df = pd.DataFrame(np.array([['XGBoost',max_depth, n_estimators, max_leaves, learning_rate, acc, roc]]),
                            columns = results_df.columns)
    results_df = pd.concat([results_df, model_df])

display(results_df)
````

````python
with open('best_mnist_xg.pkl', 'rb') as f:
    clf = pickle.load(f)
    
acc = accuracy_score(y_test, clf.predict(X_test))
roc = roc_auc_score(y_test, clf.predict_proba(X_test)[:,1])

print('Classification accuracy = {:.1f}%'.format(acc*100))
print('ROC-AUC score = {:.5f}'.format(roc))

# Plot ROC-AUC curve

fpr, tpr, thresholds = roc_curve(y_test, clf.predict_proba(X_test)[:,1])

fig, ax = plt.subplots(1,1,figsize = (6, 4), dpi = 150)
ax.plot(fpr, tpr, color='#D55E00')
ax.set_xlabel('False positive rate',fontsize = 20)
ax.set_ylabel('True positive rate',fontsize = 20)
ax.set_title('ROC curve, best XGBoost',fontsize = 24)
ax.xaxis.set_minor_locator(MultipleLocator(0.04))
ax.yaxis.set_minor_locator(MultipleLocator(0.04))
ax.tick_params(which='both',direction='in',top=True,right=True,labelsize = 16)
ax.grid(color='xkcd:dark blue',alpha = 0.2)
````
